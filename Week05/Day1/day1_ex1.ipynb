```python
import pandas as pd
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load Dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df = pd.read_csv(url)
```

```python
# Define features and target
features = df[['total_bill', 'size']]
target = df['tip']
```

```python
print("Features: \n", features.head())  # Display first few rows of features
print("Target: \n", target.head())  # Display first few rows of target
```

```python
# Split the dataset into training and testing sets
test_size = 0.2  # Proportion of the dataset to include in the test split
random_state = 42  # Set random state for reproducibility
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=test_size, random_state=random_state)
```

```python
print("Training Data Set: ", X_train.shape)  # Print shape of training data
print("Testing Data Set: ", X_test.shape)  # Print shape of testing data
```

```python
# Visualize the relationships
sns.pairplot(df, x_vars=['total_bill', 'size'], y_vars='tip', height=5, aspect=0.8, kind='scatter')
plt.title("Features vs Traget Relationship")
plt.show()  # Show the plot
```
