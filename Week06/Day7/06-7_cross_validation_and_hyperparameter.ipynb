# 交差検証（Cross-Validation）

* **交差検証とは？**

  * 機械学習モデルが独立したデータセットにどれだけ汎化できるかを評価する手法

* **交差検証の種類**

  * **K分割交差検証（K-Fold Cross-Validation）**

    * データセットをほぼ同じサイズの K 個に分割
    * モデルを K−1 個のデータで学習し、残り1つで検証
    * このプロセスを K 回繰り返し、平均性能を算出

  * **層化K分割交差検証（Stratified K-Fold）**

    * 各分割が元のデータセットと同じクラス分布を保つようにする
    * 不均衡データセットに有用

  * **Leave-One_Out 交差検証（Leave-One-Out Cross-Validation, LOOCV）**

    * 単一のデータポイントを検証用に、残りを学習用に使用
    * 全データポイントに対してこのプロセスを繰り返す
    * 計算コストは高いが、最も堅牢な評価を提供

# ハイパーパラメータチューニング

* **ハイパーパラメータチューニングとは？**

  * ハイパーパラメータはモデルによって学習されるのではなく、学習前に設定されるパラメータであり、これらを調整することはモデル性能の最適化に不可欠

* **ハイパーパラメータチューニングの手法**

  * **グリッドサーチ（Grid Search）**

    * あらかじめ定義されたハイパーパラメータ空間を網羅的に探索
    * 例：`max_depth` と `learning_rate` のすべての組み合わせをテスト
  * **ランダムサーチ（Random Search）**

    * 定義済みのパラメータ空間からハイパーパラメータの組み合わせをランダムにサンプリング
    * パラメータ空間が大きい場合、グリッドサーチより効率的

* **ハイパーパラメータチューニングの重要性**

  * 最適な設定を選択することで過学習や過小適合を防ぐ
  * 重要な設定を最適化することでモデル性能を向上させる
